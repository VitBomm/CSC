```python
# tinh cac tip la cac outlier (upper) cua nguoi nam
Q1 = tips.tip.quantile(0.25)
Q3 = tips.tip.quantile(0.75)
IQR = Q3 - Q1
print('IQR:',IQR)
upper_value = Q3 + 1.5*IQR
print('upper value:',upper_value)
```

```python
tips[tips['tip']>upper_value].sort_values(by='tip')
```
