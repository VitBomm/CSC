13. Câu 13: Thêm cột result căn cứ dữ liệu của cột 'score', nếu >=10 thì giá trị = 1

```python
df2['result'] = df2['score'].map(lambda x: 1 if x >= 10 else 0)
```

14 .Câu 14: Trong df2, thay tên 'Emily' thành 'Samantha'

15. Duyệt df2, in name, score, result: nếu = 1 thì in 'Pass', ngược lại thì in 'Fail'

```python
for index, row in df2.iterrows():
    if row['result'] == 1:
        print('name',row['name'],'score',row['score'],'attemps',row['attempts'],'qualify',row['qualify'],':','Pass')
    else:
        print('name',row['name'],'score',row['score'],'attemps',row['attempts'],'qualify',row['qualify'],':','Fail')
```
