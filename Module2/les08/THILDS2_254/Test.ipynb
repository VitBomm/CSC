```python
import numpy as np
import pandas as pd
```

```python
# Câu 1.Tạo DataFrame employees chứa danh sách các nhân viên. Danh sách này được đọc từ tập tin employees.xls
employees = pd.read_excel('employees.xls')
employees.info()
```

```python
employees.head()
```

```python
departments = pd.read_excel('departments.xls')
departments
```

```python
jobs = pd.read_excel('jobs.xls')
jobs
```

```python
emp = employees[['EMPLOYEE_ID', 'FIRST_NAME', 'LAST_NAME', 'HIRE_DATE', 'SALARY', 'DEPARTMENT_ID', 'JOB_ID']]
```

```python
emp.head()
```

```python
# 5	Trong DataFrame emp hãy thêm vào hai cột: DEPARTMENT_NAME (tên phòng) và JOB_TITLE (tên công việc)  theo mẫu sau
cau5 = pd.merge(emp, departments[['DEPARTMENT_ID','DEPARTMENT_NAME']], on = 'DEPARTMENT_ID', how='left')
cau5 = pd.merge(cau5, jobs[['JOB_ID', 'JOB_TITLE']], on = 'JOB_ID', how='left')
```

```python
cau5
```

```python
# Câu 6 Liệt kê các nhân viên của phòng có department_id là 50, sắp giảm theo salary.
cau6 = employees[employees['DEPARTMENT_ID'] == 50]
```

```python
cau6.sort_values("SALARY", ascending = False) 
```

```python
# 7 Liệt kê các nhân viên của phòng có department_name là Sales và có salary>=5000 và salary<=7000
employee_7 = pd.merge(employees, departments[['DEPARTMENT_ID', 'DEPARTMENT_NAME']], on= 'DEPARTMENT_ID', how = 'left')
employee_7 = employee_7[(employee_7['SALARY'] >= 5000) & (employee_7['SALARY'] < 7000)]
employee_7 = employee_7[employee_7['DEPARTMENT_NAME'] == 'Sales']
```

```python
employee_7
```

```python
#8 Liệt kê các nhân viên làm các công việc sao cho mã công việc (job_id) có chứa chuỗi MAN
cau8 = employees[employees['JOB_ID'].str.contains('MAN')]
cau8
```

```python
#Cau 9 Cho biết nhân viên không thuộc phòng ban nào cả
cau9 = employees[employees['DEPARTMENT_ID'].isnull()]
cau9
```

```python
# Cau 10
employees.SALARY.describe()
```

```python
#11 Liệt kê các nhân viên có lương (salary) > trung bình các lương, sắp tăng theo department_name và sắp giảm theo salary
employees11 = pd.merge(employees, departments[['DEPARTMENT_ID', 'DEPARTMENT_NAME']], on= 'DEPARTMENT_ID', how = 'left')
cau_11 = employees11[employees11['SALARY'] > employees11.SALARY.mean()]
cau_11.sort_values(by=['DEPARTMENT_NAME', 'SALARY'], ascending= [True,False])

```

```python
# 12 Cho biết nhân viên có lương lớn nhất, nhân viên có lương nhỏ nhất
cau12_1 = employees[employees['SALARY'] == employees.SALARY.min()]
cau12_2 = employees[employees['SALARY'] == employees.SALARY.max()]
```
