Cho 2 tập tin heights.txt và positions.txt => hãy chép dữ liệu từ 2 tập tin
vào vào 2 list là heights và positions
'GK' (goalkeeper), 'M' (midfield), 'A' (attack) and 'D' (defense).
1. Chuyển 2 list positions và heights thành numpy arrays: np_positions, np_heights
2. Tính median của các GK (Goal Keeper)
3. Tính median của những vị trí khác (Không phải là Goal Keeper)

```python
import numpy as np
import ast
```

```python
with open('positions.txt','r') as f:
    data = f.read()
data = data.replace("'",'')
positions = np.array(data.split(', '),dtype='U')
print('Position',positions)
print('Shape Positions',np.shape(positions))
with open('heights.txt','r') as f:
    data_h = f.read()
data_h
heights = np.array(ast.literal_eval(data_h),dtype = int)
print('Heights',heights)
print('Shape Heights', np.shape(heights))
```

```python
# 2. Tính median của các GK (Goal Keeper)
# Heights of the goalkeepers: gk_heights
condition = np.where(positions == 'GK')
print('Condition',condition)
gk_heights = heights[positions == 'GK']
# Heights of the other players: other_heights
other_heights = heights[positions != 'GK']
# Print out the median height of goalkeepers
print("Median height of goalkeepers: " + str(np.median(gk_heights)))
# Print out the median height of other players
print("Median height of other players: " + str(np.median(other_heights)))
```
